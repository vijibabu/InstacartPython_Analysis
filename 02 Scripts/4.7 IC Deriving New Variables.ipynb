# 2. if statements with user-defined functions

```python
# Create a subset

df = ords_prods_merged[:1000000]
```

```python
df.shape
```

```python
# Define a function

def price_label(row):
    
    if row['prices'] <= 5:
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High-range product'
    else: return 'np.nan'
```

```python
# Apply the function

df['price_range'] = df.apply(price_label, axis = 1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
df['prices'].max()
```

# 4. if statements with for-loop

```python
# Frequency of orders_day_of_the_week column

ords_prods_merged['orders_day_of_week'].value_counts(dropna = False)
```

```python
# Create for-loop

result = []

for value in ords_prods_merged["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

# 1. Create new column for busiest days and slowest days

```python
# Create for loop 

result_new = []

for value in ords_prods_merged["orders_day_of_week"]:
  if (value == 0) or (value ==1):
    result_new.append("Busiest days")
  elif (value == 3) or (value == 4):
    result_new.append("Slowest days")
  else:
    result_new.append("Regularly busy days")
```

```python
# Check the output

result_new
```

```python
# Creating a new column in the dataframe with the new results

ords_prods_merged['busiest_days'] = result_new
```

```python
# Check the frequency

ords_prods_merged['busiest_days'].value_counts(dropna = False)
```

# Weekends being busiest days for IC and tuesday and wednesday being the least busy days.

# 2. Create new column busiest_period_of_day

```python
# Check frequency to find the busiest hours

ords_prods_merged['order_hour_of_day'].value_counts(dropna = False)
```

```python
# Create for loop 

result_new1 = []

for value in ords_prods_merged["order_hour_of_day"]:
  if (value >= 9) and (value <= 16):
    result_new1.append("Most orders")
  elif value >= 0 and value <=6 or value == 23:
    result_new1.append("Fewest orders")
  else:
    result_new1.append("Average orders")
```

```python
result_new1
```

```python
# Creating a new column in the dataframe with the new results

ords_prods_merged['busiest_period_of_day'] = result_new1
```

```python
# Check the frequency

ords_prods_merged['busiest_period_of_day'].value_counts(dropna = False)
```

# Busiest period of the day is between 9 am - 4 pm and fewest orders are during 11pm - 6 am.
