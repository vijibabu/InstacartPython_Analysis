# 4. Creating a loyalty flag for existing customers using the transform() and loc() functions

```python
# Create a new column for max_order with grouping by user_id and using trasnform function

ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
#Check the output

ords_prods_merge.head(15)
```

```python
ords_prods_merge.head(100)
```

# Deriving column with loc()

```python
#Creating flag column

ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
#Check the frequency of newly derived column

ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
#Check the output with limited columns

ords_prods_merge[['user_id','loyalty_flag', 'order_number']].head(60)
```

# 5. Whether there’s a difference between the spending habits of the three types of customers

```python
# Group by loyalty category and aggregate to find the basic statistics

ords_prods_merge.groupby('loyalty_flag').agg({'prices' : ['mean', 'median', 'min', 'max', 'count','sum']})
```

# The mean is greater than the median in all three customer category. This means that the distribution is skewed to the right. The mean of New customer is higher compared to the mean of other customers. This might be because the price of the products purchased by new customer is higher compared to regular and loyal customers.

# 6. Create a spending flag for each user based on the average price across all their orders

```python
# Create a new column for avg_price with grouping by user_id and using trasnform function

ords_prods_merge['avg_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
#Check the output

ords_prods_merge.head(10)
```

```python
# Creating spending_flag column with criteria

ords_prods_merge.loc[ords_prods_merge['avg_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['avg_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
#Check the frequency of newly derived column

ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
#Check the output with limited columns

ords_prods_merge[['user_id','spending_flag']].head(60)
```

# 7.  Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column

```python
# Create a new column for median_days_prior_order with grouping by user_id and using trasnform function

ords_prods_merge['median_days_prior_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Check the output

ords_prods_merge.head()
```

```python
# Creating order_frequency_flag

ords_prods_merge.loc[ords_prods_merge['median_days_prior_order'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_days_prior_order'] > 10) & (ords_prods_merge['median_days_prior_order'] <= 20), 'order_frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days_prior_order'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
#Check the frequency of newly derived column

ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```
